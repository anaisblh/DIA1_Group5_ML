Problématique du projet :
Can we predict whether an employee is at risk of developing mental health issues based on workplace and lifestyle factors?

En français :
Peut-on prédire si un employé est à risque de développer des problèmes de santé mentale à partir de ses conditions de travail et de ses habitudes personnelles ?

Étape	Description	Responsable
1. Importation + EDA : Import des librairies, chargement du dataset, statistiques de base, visualisation, corrélations	Tous ensemble (travail commun)
2. Préprocessing : Nettoyage, encodage, gestion des valeurs manquantes, scaling, split	Tous ensemble (travail commun)
3. Baseline models	Modèles simples : Logistic Regression, Decision Tree, KNN, SVM	(Personne 1)
4. Hyperparameter tuning + Pipelines : GridSearchCV, cross-validation, pipeline structuré	(Personne 2)
5. Advanced models	Ensemble learning : Random Forest, XGBoost, Stacking	Personne 3
6. Deep Learning + Scientific paper	: MLP (TensorFlow/Keras), citation de l’article, comparaison avec les autres modèles	(Personne 4)
7. Model evaluation & critical analysis	: Comparaison des résultats, discussion, conclusions, visualisation finale	(Tous ensemble)

Cible du dataset :

Treatment → binaire :

Yes → la personne a déjà suivi un traitement de santé mentale.

No → la personne n’a pas suivi de traitement.

L’idée est que cette variable représente indirectement le niveau de risque ou de vulnérabilité d’un individu vis-à-vis de la santé mentale.
